```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([5,6,7,12,15,90])
y=np.array([2,65,78,42,9,10])
plt.plot(x,y)
```

```python
x1=np.array([3,4,56,7,8])
y1=np.array([5,6,7,89,8])
plt.plot(x1,y1,marker="o")
```

```python
plt.plot(x,y,x1,y1)
```

```python
plt.plot(x1,y1,"o",x,y,"o")
```

```python
plt.plot(x1,y1,marker="*",ms=10,mec="r",mfc="y") #markerfillcolor(mfc)
```

```python
plt.plot(x,y,linestyle='dashdot',color="g") #to styleline
```

```python
plt.plot(x,y,linestyle='dashed',color='m')
plt.plot(x1,y1,linestyle='dotted',color='#340012',linewidth=8)
```

```python
font={'size':15,'color':"b"}
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("demo chart",fontdict=font,loc='left')
plt.plot(x,y,linestyle='dashed',color='m')
plt.grid()#gives gripd line to your graph
```

```python
plt.plot(x,y,linestyle='dashed',color='m')
plt.grid(axis='x',linestyle='-.',color='b')
plt.subplot(1,2,1)
plt.title("d1")
plt.plot(x,y)
plt.subplot(1,2,2)
plt.title("d2")
plt.plot(x1,y1)
plt.suptitle("super title")
```
